```python
import pandas as pd
filePath=r'C:\Users\cse319\Documents\DataFiles\AppleStore.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
d = readCsvData(filePath)
print(d)
```

```python
t=d[['prime_genre','user_rating']]
g=t[t.prime_genre=='Games']
ng=t[t.prime_genre!='Games']
print("Mean {0}".format(g.mean()))
print("mean not games {0}".format(ng.mean()))
```

```python
movies=['free app','Afforable apps','Expensive app','vrey Expensive']
rating1=d[d.price==0]
rating2=d[(d.price>0)&(d.price<20)]
rating3=d[(d.price>20)&(d.price<50)]
rating4=d[d.price>50]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea=rating3.count()[0]
ve=rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=movies,shadow=True,startangle=90,autopct="%1.1f%%")
plt.show()
```
